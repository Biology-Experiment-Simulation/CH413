## The SMILES format
Our database contains 1128 lines, one line / molecule. In each line, we have the structure of the molecule (1st column) and its solubility (2nd column). The structure is encoded in the so-called [SMILES format](https://en.wikipedia.org/wiki/Simplified_molecular-input_line-entry_system), which contains a limited amount of information about the structure - chiefly the identity of the atoms and their connectivity (i.e. which bonds are helding them together). A SMILES structure is thus 2D representation of a molecule - let's have a look...

```python
# Load saved train/test sets...
# Feel free to create your own train/test splits...
Train=np.load('./ES_ML_SOSSO/Train.npy',allow_pickle = True)
Test=np.load('./ES_ML_SOSSO/Test.npy',allow_pickle = True)
TrainTargets=np.load('./ES_ML_SOSSO/TrainTargets.npy',allow_pickle = True)
TestTargets=np.load('./ES_ML_SOSSO/TestTargets.npy',allow_pickle = True)
```

```python
# Convert into a different format (rdkit internal)
MFSTR=cf.SMILES2MOLES(Train)
MFSTE=cf.SMILES2MOLES(Test)
```

```python
# Choosing the descriptors for training (SDTR) and test (SDTE) sets. 
from rdkit.Chem import Descriptors
```

```python
SDTR=[]
for m in MFSTR:
    desc=[]
    desc.append(Chem.Lipinski.NumAromaticRings(m)) # Number of aromatic rings
    desc.append(Chem.Lipinski.NumHAcceptors(m))    # Number of hydrogen bond acceptors
    desc.append(Chem.Lipinski.HeavyAtomCount(m))   # Number of heavy atoms within the molecule
    desc.append(Descriptors.TPSA(m))               # from J. Med. Chem. 43:3714-7, (2000)
    SDTR.append(desc)
  
    
SDTE=[]
for m in MFSTE:
    desc=[]
    desc.append(Chem.Lipinski.NumAromaticRings(m)) 
    desc.append(Chem.Lipinski.NumHAcceptors(m))
    desc.append(Chem.Lipinski.HeavyAtomCount(m))
    desc.append(Descriptors.TPSA(m))
    SDTE.append(desc)
```

```python
# Use a Gaussian process (GP) to fit our training set

# Choosing the kernel: radial basis function, initial lengthscale =1

kernel=RBF(length_scale_bounds=(5e-02,10)) 

gp=GaussianProcessRegressor(kernel=kernel,normalize_y=True,n_restarts_optimizer=10,alpha=1e-4);

# Fit our GP model
gp.fit(SDTR,TrainTargets);
```

```python
# Use the trained GP to predict the solubility of our training set...
TrainPred=gp.predict(SDTR).reshape(-1,1)
# ... and of our test set
TestPred=gp.predict(SDTE).reshape(-1,1)
# The reshape(-1,1) bit is needed to convert numpy arrays into scikit-learn compatible inputs...
```

```python
from sklearn.metrics import mean_squared_error

# How are we doing? 
# Let us calculate the mean squared error of our prediction for the training set 
# with respect to the actual solubility values
mse_train=mean_squared_error(TrainTargets,TrainPred)
print("MSE (training)) = ", mse_train)

# Now let's do the same for the test set...
mse_test=mean_squared_error(TestTargets,TestPred)
print("MSE (test)) = ", mse_test)
```

```python
# Let's have a look!
fig=plt.figure(num=None,figsize=(4,4),dpi=500,facecolor='w',edgecolor='k')
plt.xlabel('Exp. Solubility [log(mols/litre)]',fontsize=8)
plt.ylabel('GP prediction [log(mols/litre)]',fontsize=8)
x = np.linspace(min(TrainTargets), max(TrainTargets), 1000)
plt.plot(x, x + 0, '-r')
plt.scatter(TrainTargets,TrainPred,color='rebeccapurple',alpha=0.75,label='Training Set')
plt.scatter(TestTargets,TestPred,color='lime',alpha=0.5,label='Test Set',facecolors='none')
plt.legend(fontsize=8,loc=4);
#plt.savefig('4.png');
```

```python
plt.figure(num=None,figsize=(4,4),dpi=500,facecolor='w',edgecolor='k')
plt.xlabel('Hydration free energy [(probably) kJ/mol]',fontsize=8)
#plt.legend(fontsize=8,loc=4);
ExpTrain = np.array(TrainSet['expt']).reshape(-1,1) #Create an array set for target values
ExpTest = np.array(TestSet['expt']).reshape(-1,1)
TrainSet['expt'].plot.kde(bw_method = 0.5, label="Train") #bw_method method used to calculate the estimator bandwidth
TestSet['expt'].plot.kde(bw_method = 0.5, label="Test") #plot.kde method used to calculate the estimator bandwidth
plt.legend()
plt.show()
```

```python
MFSTR = cf.SMILES2MOLES(np.array(TrainSet['smiles']))
MFSTE = cf.SMILES2MOLES(np.array(TestSet['smiles']))
```

```python
from rdkit.Chem import Lipinski
def describe(smiles,descriptors=[Lipinski.NumHAcceptors,Lipinski.NumHDonors]):
    SD=[]
    for m in smiles:
        desc=[]
        for d in descriptors:
            desc.append(d(m))
        SD.append(desc)
    return SD
```

```python
# descriptor_list=[Lipinski.HeavyAtomCount,
#                  Lipinski.NOCount,
#                  Lipinski.NumHAcceptors,
#                  Lipinski.NumHDonors,
#                  Descriptors.MaxAbsPartialCharge,
#                  Descriptors.MinAbsPartialCharge,
#                  Descriptors.MolLogP,
#                  Descriptors.MolWt,
#                  Descriptors.NHOHCount,
#                  Descriptors.NumAromaticRings,
#                  Descriptors.NumHeteroatoms,
#                  Descriptors.TPSA]

descriptor_list=[Descriptors.MolLogP,
Lipinski.NOCount,
Descriptors.NumAromaticRings,
Lipinski.NumHAcceptors,
Lipinski.NumHDonors,  
Descriptors.TPSA,               
Descriptors.MaxAbsPartialCharge, 
Descriptors.MinAbsPartialCharge]

SDTR=describe(MFSTR,descriptor_list)
SDTE=describe(MFSTE,descriptor_list)
```

```python
from sklearn.gaussian_process import kernels
kernel = kernels.RationalQuadratic (length_scale=1,alpha=0.1) + kernels.WhiteKernel(noise_level=1)
```

```python
Gaussian = GaussianProcessRegressor(kernel=kernel, normalize_y=True, n_restarts_optimizer=10,alpha=1e-4);
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
# from sklearn.metrics import mean_squared_error
# from sklearn.model_selection import train_test_split
# from sklearn.model_selection import cross_val_score #The simplest way to use cross-validation helper function on the estimator and the dataset.
# from sklearn import metrics #scoring parameter

# from rdkit.Chem import rdMolDescriptors #Choosing the descriptors

mse_train=metrics.mean_squared_error(ExpTrain,PredictionTrain)
mse_test=metrics.mean_squared_error(ExpTest,PredictionTest)
score = np.mean(cross_val_score(gp, PredictionTest, ExpTest, cv=5, scoring = "neg_mean_squared_error",n_jobs = -1))

print("mean squared error in training set is " + str(mse_train))
print("mean squared error in test set is " + str(mse_test))

#Get a cross validation score
print("The cross validation score is " + str(score) + ".")
```

We can try optimising the coefficient of the kernel, using a constant kernel to multiply the rational quadratic kernel by, so we get a better ratio of contributions from the rational quadratic kernel to the white kernel. We can build this as a new kernel and view the updated our gaussian estimator.

```python
k2 = kernels.ConstantKernel()*kernels.RationalQuadratic(length_scale=1,alpha=0.1)+ kernels.WhiteKernel(noise_level=1)
gp.set_params(kernel = k2)
```

```python
from sklearn.model_selection import GridSearchCV
testalpharange = np.arange(0.1,0.6,0.1) # setting which values of alpha to search
testconstantrange = np.arange(2,12,2) # setting which constants to search to scale Rational Quadratic Kernel by
grid = GridSearchCV(gp,
                    param_grid={'kernel__k1__k2__alpha':testalpharange,
                                   'kernel__k1__k1__constant_value': testconstantrange}, #parameters to be tested
                    scoring ="neg_mean_squared_error",# using mean squared error as a scorer
                    cv=5, #cross validates using 5 folds. i.e uses 20% of data as test set each time, 5 times.
                    iid = True, #added as a recent update changes the default of this value, and I want results to be consistent
                    n_jobs = -1, #Computer uses all available processing power to execute this line
                    verbose = 2) #verbosity added so an idea of the progress of the calculation can be gathered



#setting up the cross validation search.
grid.fit(SDTR,ExpTrain) 

# searches for which value of alpha  and which scaler to the kernel will provide the best cross validation score 
# based on the data we have provided, and fits our estimator to that data
```

We can then use the the best parameters from the grid search to generate another set of predictions. We don't even need to extract the parameters, we can plug them straight into 'grid', and save these predictions as variables.

```python
# predicting for the training and test sets using the optimised parameters
NewTrainPred=grid.predict(SDTR).reshape(-1,1) 
NewTestPred=grid.predict(SDTE).reshape(-1,1)
```

```python
# calculating the mean squared error for the new plot
new_mse_train=metrics.mean_squared_error(ExpTrain,NewTrainPred)
new_mse_test=metrics.mean_squared_error(ExpTest,NewTestPred)
print("The new MSE error in training set is " + str(mse_train))
print("The new MSE error in test set is " + str(mse_test))
new_score = np.mean(grid.best_score_)
#Get a cross validation score
print("The new cross validation score is " + str(new_score))
```

```python
# plotting the data
fig=plt.figure(num=None,figsize=(3,3),dpi=500,facecolor='w',edgecolor='k')
plt.xlabel('Exp. Hydration free energy [(probably) kJ/mol]')
plt.ylabel('Predicted Hydration free energy [(probably) kJ/mol]')
x = np.linspace(min(NewTrainPred), max(NewTrainPred), 1000)
plt.plot(x, x + 0, '-r')
plt.scatter(ExpTrain,NewTrainPred,alpha=0.5,label='Training Set', marker = '.')
plt.errorbar(ExpTrain, NewTrainPred, yerr = np.sqrt(-new_score),  ls = 'none' )
plt.scatter(ExpTest,NewTestPred,alpha=0.5,label='Test Set', marker = '.')
plt.errorbar(ExpTest, NewTestPred, yerr = np.sqrt(-new_score),  ls = 'none', alpha = 0.6)
plt.legend(fontsize=8,loc=2);
```
